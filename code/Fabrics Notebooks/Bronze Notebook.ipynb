# Bronze Notebook
### This is used to ingest data into the bronze layer in the berka lakehouse

```python
import requests
import os
import zipfile

#url for zipped file
## derived from Kaggle API and actual url used to avoid need to install Kaggle library for the person trying to reproduce

url = "https://www.kaggle.com/api/v1/datasets/download/marceloventura/the-berka-dataset?dataset_version_number=1"

# Ensure folder exists
folder_path = "/lakehouse/default/Files/bronze_layer"
os.makedirs(folder_path, exist_ok=True)

save_path = os.path.join(folder_path, "berka.zip")

# Download
r = requests.get(url, allow_redirects=True)

# Write to lakehouse
with open(save_path, "wb") as f:
    f.write(r.content)

print("Saved to:", save_path)
```

```python
## Unzip file

import zipfile

with zipfile.ZipFile("/lakehouse/default/Files/bronze_layer/berka.zip") as z:
    z.extractall("/lakehouse/default/Files/bronze_layer")

print("Extracted all files")
```

```python
## I removed zipped file here but in practice it is moved to an archived file for audit purposes.

os.remove("/lakehouse/default/Files/bronze_layer/berka.zip")
print("Zip deleted")
```
